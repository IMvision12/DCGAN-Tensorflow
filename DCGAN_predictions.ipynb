```python
import tensorflow as tf
import numpy as np
import cv2

import matplotlib.pyplot as plt
```

```python
model = tf.keras.models.load_model("gen.h5",compile=False)
```

```python
def denormalize(x):
    return (x + 1) * 127.5
```

```python
def predict(model):
    noise = tf.random.normal([1, 100])
    pred = model.predict(noise)
    pred = np.squeeze(pred)
    pred = denormalize(pred)
    return pred
```

```python
noise = tf.random.normal([1, 100])

_,ax = plt.subplots(5,5, figsize = (8,8))
for i in range(5):
    for j in range(5):
        img = predict(model)
        ax[i,j].imshow(img.astype("uint8"))
        ax[i,j].axis('off')
```
